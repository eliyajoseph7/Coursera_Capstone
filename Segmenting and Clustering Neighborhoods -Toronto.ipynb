## Using BeautifulSoup to scrape the wikipedia page and get the table

```python
source = requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M').text
soup=BeautifulSoup(source,'lxml')
print(soup.title)
from IPython.display import display_html
my_table = str(soup.table)
display_html(my_table,raw=True)
```

## Converting the html table to Pandas dataFrame

```python

dfs = pd.read_html(my_table)
df=dfs[0]
df.head()
```

## Data preprocessing and cleaning


```python
# Dropping the rows where Borough is 'Not assigned'
df1 = df[df.Borough != 'Not assigned']
df1.rename(columns={'Postal Code':'PostalCode'},inplace=True)

# Combining the neighbourhoods with same Postalcode
df2 = df1.groupby(['PostalCode','Borough'], sort=False).agg(', '.join)
df2.reset_index(inplace=True)

# Replacing the name of the neighbourhoods which are 'Not assigned' with names of Borough
df2['Neighbourhood'] = np.where(df2['Neighbourhood'] == 'Not assigned',df2['Borough'], df2['Neighbourhood'])

df2.head()
```

```python
df2.shape
```

### Importing the csv file conatining the latitudes and longitudes for various neighbourhoods in Canada

```python
lat_lon = pd.read_csv('https://cocl.us/Geospatial_data')
lat_lon.head()
```

## Merging two dataframe tables to obtain the longitudes and latitudes of the neighbours

```python
lat_lon.rename(columns={'Postal Code':'PostalCode'},inplace=True)
df3 = pd.merge(df2,lat_lon,on='PostalCode')
df3.head()
```

## Getting all rows which contains the word Toronto in them

```python
toronto = df3[df3.Borough.str.contains('Toronto', regex=False)]
toronto.head()
```

```python
address = 'Toronto, ON, Canada'
geolocator = Nominatim(user_agent='the_toronto')
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude

print("The latitude and longintude of Toronto are {} and {} respectively".format(latitude, longitude))
```

```python
map_tor = folium.Map(location=[latitude, longitude], zoom_start=10)

# Add markers to the map
for lat, lon, borough, neighbourhood in zip(toronto['Latitude'], toronto['Longitude'], toronto['Borough'], toronto['Neighbourhood']):
    labels = '{}, {}'.format(neighbourhood, borough)
    labels = folium.Popup(labels, parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        color='red',
        fill_color = 'blue',
        radius = 5,
        popup=labels,
        fill_opacity=0.7,
        fill=True,
        parse_html=False
    ).add_to(map_tor)
map_tor
```
